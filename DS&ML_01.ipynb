```python
import matplotlib.pyplot as plt
```

```python
sub_dataset1.groupby('location')['new_cases'].count()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14,14))
sub_dataset1.groupby('location')['new_cases'].plot(ax=axes[0,0], legend=True)
sub_dataset1.groupby('location')['new_deaths'].plot(ax=axes[0,1], legend=True)
sub_dataset1.groupby('location')['total_cases'].plot(ax=axes[1,0], legend=True)
sub_dataset1.groupby('location')['total_deaths'].plot(ax=axes[1,1], legend=True)

axes[0,0].set_title("New Cases")
axes[0,1].set_title("New Deaths")
axes[1,0].set_title("Total Cases")
axes[1,1].set_title("Total Deaths")
```
